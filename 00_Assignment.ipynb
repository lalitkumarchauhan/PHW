The "WHILE LOOP" can be used where the information about total number of iteration is not pre-known or some specific output is deemed to be the facilitator to stop the execution. In the below example the execution is stopped when the condition is False. identifier a is initialised to 3 and "sent" flag is set to True initally. Now, the while is attached to the condition where "sent" is True only so the execution will enter the while loop if "sent" is True. At first step "sent" is True, so the execution enter to if statement and since a is 3 the subsequent steps are executed and 3 is printed and "sent" is set to True, else will execute only if the "if" condition fails, so next, the value of a is incremented by 2 and a becomes 5 now since the "sent" flag is True, the if condition is encountered once more and since the a is 5 and "if" will not execute as a must be less than 5 and not equal to 5, so execution moves ahead to "else" where "sent" is set to "False" and the a is set to 7 now, moving forward the While will check the value of "sent" and finds the value of the flag to be False, so the execution stops. Please note here the number of iterations were not known previously hence for loop is not applicable.

```python
a = 3
sent = True
while sent == True:
    if a >= 3 and a < 5:
        print(a)
        sent = True
    else:
        sent = False
    a = a + 2
```

```python
n = int(input("Enter the number of rows you want in the right skewed pyramid pattern: "))
i = 1
j = 0
while i <= n:
    while j <= i - 1:
        print('* ', end = "")
        j = j + 1
    print("\n")
    i = i + 1
    j = 0
```

```python
n = int(input("Enter the number of rows you want in the pyramid pattern: "))
i = n
j = 0
while i >= 0:
    while j <= i - 1:
        print('* ', end = "")
        j = j + 1
    print("\n")
    i = i - 1
    j = 0
```
